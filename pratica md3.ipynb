```python
num1 = int( input('Adicione o numero: '))
num2 = int( input('Adicione o numero: '))
opr = input('qual operação você deseja fazer? (+,-,*,/)')

if opr == "+" or opr == 'somar':
  res = num1 + num2
elif opr == '-' or opr == 'subtrair':
  res = num1 - num2
elif opr == '*' or opr == 'multiplicar':
  res = num1 * num2
elif opr == '/' or opr == 'dividir':
  res = num1 / num2
else:
  res = 'Operação invalida'

print('Resultado: ', res)
```

```python
while True:
  print('\n=== CALCULADORA ===')

  num1 = int( input('Adicione o numero: '))
  num2 = int( input('Adicione o numero: '))
  opr = input('qual operação você deseja fazer? (+,-,*,/)')
  if opr == 'sair':
      print('Encerrando...')
      break
  elif opr == "+" or opr == 'somar':
       res = num1 + num2
  elif opr == '-' or opr == 'subtrair':
      res = num1 - num2
  elif opr == '*' or opr == 'multiplicar':
      res = num1 * num2
  elif opr == '/' or opr == 'dividir':
      res = num1 / num2
  else:
      print('Operação invalida')
      continue

  print('Resultado: ', res)
```
